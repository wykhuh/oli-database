```python
import pandas as pd
```

```python
def log_df(df, limit=5):
    print(df.shape)
    return df.head(limit)
```

# Merge csvs

```python
playlist_df = pd.read_csv('../raw_data/oli_playlist.csv')
log_df(playlist_df)
```

```python
oli_df = pd.read_csv('../raw_data/Oli.csv')

log_df(oli_df)
```

```python

missing_df = pd.read_csv('../processed_data/missing_videos_fixed.csv')
log_df(missing_df)
```

combine videos from 'Oli playtlist and videos missing from playlist

```python
yt_df = pd.concat([playlist_df, missing_df[playlist_df.columns]])
log_df(yt_df)
```

merge youtube data with my custom oli csv

```python
merge_df = yt_df.merge(oli_df, left_on='position', right_on='id', how='left')
log_df(merge_df)
```

add other columns to missing videos

```python
cols = [
    'tier', 'type', 'size', 'top_wood', 'back_wood', 'finish', 'other',
    'price', 'url'
]
for i, row in missing_df.iterrows():
    for col in cols:
        merge_df.loc[merge_df['video_id'] == row['video_id'], col] = row[col]

```

add model and model wood columns

```python
merge_df['new_model'] = pd.NA
merge_df['model'] = merge_df['tier'] + '-' + merge_df['type']

merge_df['new_model_wood'] = pd.NA
merge_df['model_wood'] = merge_df['tier'] + '-' + merge_df['type'].replace(r'^[BTCS]', '', regex=True)

log_df(merge_df)
```

```python
models = set()
tonewoods = set()

for i, row in merge_df.iterrows():
    if row['model'] not in models and pd.notna(row['model']):
        models.add(row['model'])
        merge_df.at[i, 'new_model'] = True

    if row['model_wood'] not in tonewoods and pd.notna(row['model_wood']):
        tonewoods.add(row['model_wood'])
        merge_df.at[i, 'new_model_wood'] = True
```

```python
log_df(merge_df)
```

```python
merge_df.to_csv('../processed_data/draft/oli_lists_temp.csv', index=False)
```

```python
final_cols = [
    'position', 'video_title', 'published_at', 'video_id', 'video_thumbnail',
    'model', 'model_wood', 'tier', 'type', 'size', 'top_wood', 'back_wood',
    'finish', 'other', 'price', 'url', 'new_model', 'new_model_wood']

final_df = merge_df[final_cols]
```
